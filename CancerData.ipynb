## Model Training 

```python
X = df_cancer.drop(columns='target')
```

```python
len(X.columns.tolist())
```

```python
y = df_cancer['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

## Support Vector Machine Model

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

## Evaluate the Model

```python
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot=True)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_predict)
```

## Improving the Model 

### Normalizing the features

```python
min_train = X_train.min()
```

```python
range_train = (X_train-min_train).max()
```

```python
X_train_scaled=(X_train-min_train)/range_train
```

```python
sns.scatterplot(X_train['mean area'], X_train['mean smoothness'], hue=y_train)
```

```python
sns.scatterplot(X_train_scaled['mean area'], X_train_scaled['mean smoothness'], hue=y_train)
```

```python
min_test=X_test.min()
range_test=(X_test-min_test).max()
X_test_scaled=(X_test-min_test)/range_test
```

```python
svc_model.fit(X_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
```

```python
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot=True)
```

```python
accuracy_score(y_test, y_predict)
```

```python
print(classification_report(y_test, y_predict))
```

## Parameters C, Gamma tuning 

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[0.1, 1, 10, 100], 'gamma':[1, 0.1, 0.01, 0.001], 'kernel':['rbf']}
```

```python
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=4)
```

```python
grid
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
grid_predict = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, grid_predict)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, grid_predict))
```
